```python
import numpy as np

# Define a function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in Length and Weight
outliers_length = detect_outliers_iqr(df_healthy, 'Length')
outliers_weight = detect_outliers_iqr(df_healthy, 'Weight')

# Print the outliers
print("Outliers in Length:")
print(outliers_length)

print("\nOutliers in Weight:")
print(outliers_weight)
```

```python
import os
import glob
import torch
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from PIL import Image
import torch.nn as nn
import torch.optim as optim
import albumentations as A
from albumentations.pytorch import ToTensorV2

# --------------------- DEVICE SETUP --------------------- #
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print("Using device:", device)

# --------------------- DATASET PREPARATION --------------------- #
base_dir = "/Users/carlosjimenez/Dropbox/Mac/Desktop/us_project/Polito-Radboud-DeepLearningUS/BB"
healthy_images = sorted(glob.glob(os.path.join(base_dir, "Healthy/Images/*.png")))
pathological_images = sorted(glob.glob(os.path.join(base_dir, "Pathological/Images/*.png")))
healthy_masks = sorted(glob.glob(os.path.join(base_dir, "Healthy/Masks/*.png")))
pathological_masks = sorted(glob.glob(os.path.join(base_dir, "Pathological/Masks/*.png")))

all_images = healthy_images + pathological_images
all_masks = healthy_masks + pathological_masks

print(f"Found {len(healthy_images)} healthy images and {len(pathological_images)} pathological images.")
print(f"Total images: {len(all_images)}")

# --------------------- DATA AUGMENTATION --------------------- #
augmentations = A.Compose([
    A.Resize(256, 256),
    A.RandomResizedCrop((224, 224), scale=(0.8, 1.0), p=0.3),  # ✅ Corrected `size`
    A.Resize(256, 256),
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=15, p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.ElasticTransform(p=0.2),
    ToTensorV2(),
])

base_transform = A.Compose([
    A.Resize(256, 256),
    ToTensorV2(),
])

# --------------------- CUSTOM DATASET CLASS --------------------- #
class UltrasoundSegmentationDataset(Dataset):
    def __init__(self, image_paths, mask_paths, augment=False):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.augment = augment

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        mask_path = self.mask_paths[idx]

        image = np.array(Image.open(img_path).convert("RGB"), dtype=np.uint8)
        mask = np.array(Image.open(mask_path).convert("L"), dtype=np.uint8)

        mask = mask / 255.0  
        mask = (mask > 0.5).astype(np.float32)  

        if self.augment:
            augmented = augmentations(image=image, mask=mask)
        else:
            augmented = base_transform(image=image, mask=mask)

        image = augmented["image"].float()
        mask = augmented["mask"].unsqueeze(0).float()

        return image, mask

# --------------------- UNET MODEL --------------------- #
class UNet(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super(UNet, self).__init__()

        def conv_block(in_channels, out_channels):
            return nn.Sequential(
                nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
                nn.ReLU(),
                nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
                nn.ReLU(),
            )

        self.encoder1 = conv_block(in_channels, 64)
        self.encoder2 = conv_block(64, 128)
        self.encoder3 = conv_block(128, 256)
        self.encoder4 = conv_block(256, 512)

        self.pool = nn.MaxPool2d(2, 2)

        self.bottleneck = conv_block(512, 1024)

        self.upconv4 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.decoder4 = conv_block(1024, 512)
        self.upconv3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder3 = conv_block(512, 256)
        self.upconv2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder2 = conv_block(256, 128)
        self.upconv1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder1 = conv_block(128, 64)

        self.final_conv = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x):
        enc1 = self.encoder1(x)
        enc2 = self.encoder2(self.pool(enc1))
        enc3 = self.encoder3(self.pool(enc2))
        enc4 = self.encoder4(self.pool(enc3))

        bottleneck = self.bottleneck(self.pool(enc4))

        dec4 = self.upconv4(bottleneck)
        dec4 = torch.cat((dec4, enc4), dim=1)
        dec4 = self.decoder4(dec4)

        dec3 = self.upconv3(dec4)
        dec3 = torch.cat((dec3, enc3), dim=1)
        dec3 = self.decoder3(dec3)

        dec2 = self.upconv2(dec3)
        dec2 = torch.cat((dec2, enc2), dim=1)
        dec2 = self.decoder2(dec2)

        dec1 = self.upconv1(dec2)
        dec1 = torch.cat((dec1, enc1), dim=1)
        dec1 = self.decoder1(dec1)

        return self.final_conv(dec1)

# --------------------- TRAINING --------------------- #
train_imgs, test_imgs, train_masks, test_masks = train_test_split(all_images, all_masks, test_size=0.15, random_state=42)
train_imgs, val_imgs, train_masks, val_masks = train_test_split(train_imgs, train_masks, test_size=0.2, random_state=42)

train_dataset = UltrasoundSegmentationDataset(train_imgs, train_masks, augment=True)
val_dataset = UltrasoundSegmentationDataset(val_imgs, val_masks, augment=False)
test_dataset = UltrasoundSegmentationDataset(test_imgs, test_masks, augment=False)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

model = UNet().to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0008)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', patience=5, factor=0.5)

best_loss = float("inf")
early_stopping_counter = 0
num_epochs = 50

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0
    for images, masks in train_loader:
        images, masks = images.to(device), masks.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_loss = epoch_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")

    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model.state_dict(), "unet_best_model.pth")
        early_stopping_counter = 0
        print(f"✅ Best model saved at epoch {epoch+1} with loss {best_loss:.4f}")
    else:
        early_stopping_counter += 1

    scheduler.step(avg_loss)

    if early_stopping_counter == 5:
        print("🚨 Early stopping triggered!")
        break

print("🚀 Model training complete.")

# --------------------- TESTING --------------------- #
model.load_state_dict(torch.load("unet_best_model.pth", map_location=device))
model.eval()

with torch.no_grad():
    total_dice = 0
    num_samples = 0
    for images, masks in test_loader:
        images, masks = images.to(device), masks.to(device)
        preds = torch.sigmoid(model(images)) > 0.5
        total_dice += (2. * (preds * masks).sum()) / ((preds + masks).sum() + 1e-8)
        num_samples += 1

print(f"✅ Final Test Set Dice Score: {total_dice / num_samples:.4f}")
```

```python
import torch
import numpy as np

# Function to calculate IoU, Precision, Recall, and Dice Score
def calculate_metrics(pred, mask):
    pred = pred > 0.5  # Convert to binary
    mask = mask > 0.5

    intersection = (pred & mask).sum()
    union = (pred | mask).sum()
    true_positive = intersection
    false_positive = pred.sum() - true_positive
    false_negative = mask.sum() - true_positive

    iou = intersection / (union + 1e-8)  # Avoid division by zero
    precision = true_positive / (true_positive + false_positive + 1e-8)
    recall = true_positive / (true_positive + false_negative + 1e-8)
    dice = (2 * intersection) / (pred.sum() + mask.sum() + 1e-8)

    return iou, precision, recall, dice

# Evaluate on test set
model.eval()
total_iou, total_precision, total_recall, total_dice = 0, 0, 0, 0
num_samples = 0

with torch.no_grad():
    for images, masks in test_loader:
        images, masks = images.to(device), masks.to(device)
        preds = torch.sigmoid(model(images))  # Get probability map
        preds = preds.cpu().numpy()
        masks = masks.cpu().numpy()

        for i in range(images.size(0)):
            iou, precision, recall, dice = calculate_metrics(preds[i, 0], masks[i, 0])
            total_iou += iou
            total_precision += precision
            total_recall += recall
            total_dice += dice
            num_samples += 1

# Compute average metrics
avg_iou = total_iou / num_samples
avg_precision = total_precision / num_samples
avg_recall = total_recall / num_samples
avg_dice = total_dice / num_samples

# Print results
print(f"✅ IoU: {avg_iou:.4f}")
print(f"✅ Precision: {avg_precision:.4f}")
print(f"✅ Recall: {avg_recall:.4f}")
print(f"✅ Dice Score: {avg_dice:.4f}")  # Already calculated before
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Ensure the model is in evaluation mode
model.eval()

# Initialize lists for storing performance metrics and images
worst_cases = []  # Stores (IoU, image, mask, prediction)
metrics = []  # Stores (IoU, Dice, Precision, Recall)

with torch.no_grad():
    for images, masks in test_loader:
        images, masks = images.to(device), masks.to(device)
        
        # Get model predictions
        preds = torch.sigmoid(model(images)) > 0.5  # Convert to binary mask

        # Convert to numpy for analysis
        images_np = images.cpu().numpy()
        masks_np = masks.cpu().numpy()
        preds_np = preds.cpu().numpy()

        for i in range(images.shape[0]):  # Loop through batch
            intersection = np.logical_and(preds_np[i], masks_np[i]).sum()
            union = np.logical_or(preds_np[i], masks_np[i]).sum()
            true_positive = intersection
            false_positive = preds_np[i].sum() - intersection
            false_negative = masks_np[i].sum() - intersection

            # Compute IoU, Dice Score, Precision, Recall
            iou = intersection / (union + 1e-8)
            dice = (2 * intersection) / (preds_np[i].sum() + masks_np[i].sum() + 1e-8)
            precision = true_positive / (true_positive + false_positive + 1e-8)
            recall = true_positive / (true_positive + false_negative + 1e-8)

            # Store the worst cases based on IoU
            metrics.append((iou, dice, precision, recall, images_np[i], masks_np[i], preds_np[i]))
            worst_cases.append((iou, images_np[i], masks_np[i], preds_np[i]))

# Sort the worst cases based on IoU
worst_cases.sort(key=lambda x: x[0])  # Sort by IoU (lowest first)

# --------------------- PLOTTING THE WORST CASES (FIXED) --------------------- #
def visualize_failure_cases(worst_cases, num_cases=5):
    fig, axes = plt.subplots(num_cases, 4, figsize=(12, num_cases * 3))

    for idx, (iou, image, mask, pred) in enumerate(worst_cases[:num_cases]):
        error_map = np.abs(mask - pred)  # Difference between ground truth and prediction

        # Ensure correct image shape
        if image.shape[0] == 3:  # If channel-first format (C, H, W), convert to (H, W, C)
            image = np.transpose(image, (1, 2, 0))

        # Convert grayscale images to RGB (matplotlib expects H, W, C)
        if len(image.shape) == 2:
            image = np.stack([image] * 3, axis=-1)

        axes[idx, 0].imshow(image.astype(np.uint8))  # Ensure uint8 format
        axes[idx, 0].set_title("Original Image")
        axes[idx, 0].axis("off")

        axes[idx, 1].imshow(mask.squeeze(), cmap="gray")  # Ground Truth Mask
        axes[idx, 1].set_title("Ground Truth Mask")
        axes[idx, 1].axis("off")

        axes[idx, 2].imshow(pred.squeeze(), cmap="gray")  # Model Prediction
        axes[idx, 2].set_title(f"Prediction (IoU: {iou:.3f})")
        axes[idx, 2].axis("off")

        axes[idx, 3].imshow(error_map.squeeze(), cmap="hot")  # Error Map
        axes[idx, 3].set_title("Error Map")
        axes[idx, 3].axis("off")

    plt.tight_layout()
    plt.show()

# Visualize the 5 worst predictions (Run this again after fixing)
visualize_failure_cases(worst_cases, num_cases=5)
```
